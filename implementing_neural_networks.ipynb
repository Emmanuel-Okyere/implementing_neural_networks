```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, Normalizer
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.optimizers import Adam
```

```python
dataset = pd.read_csv("life_expectancy.csv")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset = dataset.drop(columns = ["Country"], axis = 1)
dataset.head()
```

```python
dataset.shape
```

```python
labels = dataset.iloc[:,-1]
labels.head()
```

```python
features = dataset.iloc[:,0:-1]
```

```python
features = pd.get_dummies(features)
features_train, features_test, labels_train, labels_test = train_test_split(features, labels, train_size = 0.8, random_state = 1)
```

```python
numerical_features = features.select_dtypes(include=['float64', 'int64'])
numerical_columns = numerical_features.columns
```

```python
ct = ColumnTransformer([("only numeric", StandardScaler(), numerical_columns)], remainder='passthrough')

features_train_scaled = ct.fit_transform(features_train)
features_test_scaled = ct.fit_transform(features_test)
```

```python
my_model = Sequential()
my_model.add(layers.InputLayer(input_shape =dataset.shape[1],))
my_model.add(layers.Dense(16, activation = "relu"))
my_model.add(layers.Dense(1))
my_model.summary()
```

```python
opt = Adam(learning_rate = 0.01)
my_model.compile(optimizer = opt, metrics = ["mae"], loss = "mse")
```

```python
my_model.fit(features_train_scaled, labels_train, epochs = 40, batch_size = 1, verbose = 1)
```

```python
res_mse,res_mae = my_model.evaluate(features_test_scaled,labels_test )
```

```python
print(res_mse, res_mae)
```
